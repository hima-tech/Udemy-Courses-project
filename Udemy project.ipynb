## What is the Max. Number of Subscribers for Each Level of courses 

```python
level_of_course = df[['level', 'num_subscribers']]
lvl = level_of_course.groupby(['level'], as_index = False)['num_subscribers'].max()
```

```python
lvl_graph = lvl.sort_values(by = ['num_subscribers'], ascending = False)
lvl_graph
```

```python
clr0 = plt.bar(lvl_graph['level'],lvl_graph['num_subscribers'], color= 'gray')
clr0[0].set_color('red')
plt.title('Levels number of subscribers', fontsize = 20)
plt.xlabel('Levels', fontsize = 15)
plt.ylabel('Number Of Subscribers', fontsize = 15)
plt.tight_layout()
plt.rcParams["figure.figsize"] = (8,5)
plt.show()
```
